4. Есть список файлов  Файлы двух разных форматов. Надо их разделить по форматам на разные списки.

```python
file_list = ['a.svg', 'b.svg', 'c.tif', 'd.tif', 'e.tif']
list_tif = []
list_csv = []
for i in range(0, len(file_list)):
    if file_list[i][-3:] == 'tif':
        list_tif.append(file_list [i])
    else:
        list_csv.append(file_list [i])
    i+=1
print('list_tif :', list_tif)
print('list_csv :',list_csv)
```
